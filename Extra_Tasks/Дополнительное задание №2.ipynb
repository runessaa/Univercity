```python
phone_book = {}

def add_contact(phone_book):
  new_phone_book = phone_book.copy()
  print("Введите название для контакта: ")
  name = input().strip()
  formatted_name = ' '.join(part.capitalize() for part in name.split())

  while True:
      print("Введите номер телефона: ")
      phone = input().strip()
      cleaned_phone = ''.join(filter(str.isdigit, phone))

      if len(cleaned_phone) == 10 and cleaned_phone[0] == '9':
          new_phone_book[formatted_name] = '+79' + cleaned_phone[1:]
          print(f"Контакт {formatted_name} успешно добавлен!\n")
          break
      elif len(cleaned_phone) == 11 and cleaned_phone[0] == '7':
          new_phone_book[formatted_name] = '+7' + cleaned_phone[1:]
          print(f"Контакт {formatted_name} успешно добавлен!\n")
          break
      elif len(cleaned_phone) == 12 and (cleaned_phone.startswith('79') or cleaned_phone.startswith('7')):
          new_phone_book[formatted_name] = '+' + cleaned_phone
          print(f"Контакт {formatted_name} успешно добавлен!\n")
          break
      else:
          print("Неправильно набран номер.\n")

  return new_phone_book

def delete_contact(phone_book):
  new_phone_book = phone_book.copy()
  print("Введите название контакта для удаления: ")
  name = input().strip()
  formatted_name = ' '.join(part.capitalize() for part in name.split())

  if formatted_name in new_phone_book:
      new_phone_book.pop(formatted_name)
      print(f"Контакт {formatted_name} успешно удален.\n")
  else:
      print("Контакт не найден.\n")

  return new_phone_book

def show_contacts(phone_book):
  if phone_book:
      for name, number in phone_book.items():
          print(name, number)
  else:
      print('Нет контактов.\n')

def change_number(phone_book):
  new_phone_book = phone_book.copy()
  print("Введите название контакта: ")
  name = input().strip()
  formatted_name = ' '.join(part.capitalize() for part in name.split())

  if formatted_name in new_phone_book:
      while True:
          print("Введите новый номер телефона: ")
          new_phone = input().strip()
          cleaned_phone = ''.join(filter(str.isdigit, new_phone))

          if len(cleaned_phone) == 10 and cleaned_phone[0] == '9':
              new_phone_book[formatted_name] = '+79' + cleaned_phone[1:]
              print(f'Контакт {formatted_name} успешно изменен!\n')
              break
          elif len(cleaned_phone) == 11 and cleaned_phone[0] == '7':
              new_phone_book[formatted_name] = '+7' + cleaned_phone[1:]
              print(f'Контакт {formatted_name} успешно изменен!\n')
              break
          elif len(cleaned_phone) == 12 and (cleaned_phone.startswith('79') or cleaned_phone.startswith('7')):
              new_phone_book[formatted_name] = '+' + cleaned_phone
              print(f'Контакт {formatted_name} успешно изменен!\n')
              break
          else:
              print("Неправильно набран номер.\n")
  else:
      print("Контакт не найден.\n")

  return new_phone_book

def main():
  current_phone_book = phone_book
  print("Введите номер функции: ")
  print('1. Добавить контакт\n2. Удалить контакт\n3. Показать список контактов\n4. Изменить номер\n5. Выход')

  while True:
      try:
          print("Введите номер функции: ")
          choice = int(input())

          if choice == 1:
              current_phone_book = add_contact(current_phone_book)
          elif choice == 2:
              current_phone_book = delete_contact(current_phone_book)
          elif choice == 3:
              show_contacts(current_phone_book)
          elif choice == 4:
              current_phone_book = change_number(current_phone_book)
          elif choice == 5:
              print("Программа завершена.")
              break
          else:
              print("Неверный выбор. Попробуйте снова.")

      except ValueError:
          print("Введите число.")

if __name__ == "__main__":
  main()
```
