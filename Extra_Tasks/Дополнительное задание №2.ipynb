```python
phone_book = {}
print("Введите номер функции: ")
print('1. Добавить контакт\n2. Удалить контакт\n3. Показать список контактов\n4. Изменить номер\n5. Выход')
f = True

while f:
    try:
        print("Введите номер функции: ")
        i = int(input())

        if i == 1:
            print("Введите название для контакта: ")
            s = input().strip()
            p = ' '.join(j.capitalize() for j in s.split())
            while True:
                print("Введите номер телефона: ")
                m = input().strip()
                m_cleaned = ''.join(filter(str.isdigit, m))
                if len(m_cleaned) == 10 and m_cleaned[0] == '9':
                    phone_book[p] = '+79' + m_cleaned[1:]
                    print(f"Контакт {p} успешно добавлен!\n")
                    break
                elif len(m_cleaned) == 11 and m_cleaned[0] == '7':
                    phone_book[p] = '+7' + m_cleaned[1:]
                    print(f"Контакт {p} успешно добавлен!\n")
                    break
                elif len(m_cleaned) == 12 and (m_cleaned.startswith('79') or m_cleaned.startswith('7')):
                    phone_book[p] = '+' + m_cleaned
                    print(f"Контакт {p} успешно добавлен!\n")
                    break
                else:
                    print("Неправильно набран номер.\n")

        elif i == 2:
            print("Введите название контакта для удаления: ")
            s = input().strip()
            p = ' '.join(j.capitalize() for j in s.split())
            if p in phone_book:
                phone_book.pop(p)
                print(f"Контакт {p} успешно удален!\n")
            else:
                print("Контакт не найден\n")

        elif i == 3:
            if phone_book:
                for name, number in phone_book.items():
                    print(name, number)
            else:
                print('Нет контактов\n')

        elif i == 4:
            print("Введите название контакта: ")
            s = input().strip()
            p = ' '.join(j.capitalize() for j in s.split())
            if p in phone_book:
                while True:
                    print("Введите новый номер телефона: ")
                    m = input().strip()
                    m_cleaned = ''.join(filter(str.isdigit, m))
                    if len(m_cleaned) == 10 and m_cleaned[0] == '9':
                        phone_book[p] = '+79' + m_cleaned[1:]
                        print(f'Контакт {p} успешно изменен!\n')
                        break
                    elif len(m_cleaned) == 11 and m_cleaned[0] == '7':
                        phone_book[p] = '+7' + m_cleaned[1:]
                        print(f'Контакт {p} успешно изменен!\n')
                        break
                    elif len(m_cleaned) == 12 and (m_cleaned.startswith('79') or m_cleaned.startswith('7')):
                        phone_book[p] = '+' + m_cleaned
                        print(f'Контакт {p} успешно изменен!\n')
                        break
                    else:
                        print("Неправильно набран номер.\n")
            else:
                print("Контакт не найден\n")

        elif i == 5:
            print("Программа завершена.")
            f = False

        else:
            print("Неверный выбор. Попробуйте снова.")

    except ValueError:
        print("Введите число.")
```
